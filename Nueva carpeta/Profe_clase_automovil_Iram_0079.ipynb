### Clase automovil con 3 atributos funciones leerdats>(), mostrardatos()
#### Que sea interactivo --> los input
#### Crear 3 objetos carro1, carro2, carro3
#### No usar while True, if, for, listas ni diccionarios

![](infoautomovil.png)

```python
#clase automovil, atributos, marca, modelo, placa, precio
#funciones leerdatos(), mostardatos()
#crear 3 objetos carro1, carro2, carro3

class Automovil:
    def __init__(self, marca, modelo, placa, precio):
        self.marca = marca
        self.modelo = modelo
        self.placa = placa
        self.precio = precio

    def leerdatos(self):

        self.marca = input("Ingrese la marca del automovil: ")
        self.modelo = input("Ingrese el modelo del automovil: ")
        self.placa = input("Ingrese la placa del automovil: ")
        self.precio = float(input("Ingrese el precio del automovil: "))

    def mostrardatos(self):
        print("Marca: ", self.marca)
        print("Modelo: ", self.modelo)
        print("Placa: ", self.placa)
        print("Precio: ", self.precio)
#Fin de la clase automovil

#Creando los objetos carro1, carro2, carro3
carro1 = Automovil("", "", "", 0.0)
carro2 = Automovil("", "", "", 0.0)
carro3 = Automovil("", "", "", 0.0)

print("-"*50)
print("#Carro numero 1")
carro1.leerdatos()
carro1.mostrardatos()
print("-"*50)

print("#Carro numero 2")
carro2.leerdatos()
carro2.mostrardatos()
print("-"*50)

print("#Carro numero 3")
carro3.leerdatos()
carro3.mostrardatos()
print("-"*50)
print("Programa realizado por Iram Jimenez NL = 28")
```
